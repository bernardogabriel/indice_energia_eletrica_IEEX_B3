Margem EBIT (EBIT/Rec. Líquida) um indicador que mede a lucratividade da empresa sem considerar o efeito de juros e impostos

A Empresa tem uma Margem EBIT de X %, o que significa que X % de sua receita se transforma em lucro operacional antes de despesas financeiras e impostos

ROE Isso significa que a empresa está gerando um retorno de 20% sobre o patrimônio líquido. %

# Categoria Patrimônio 

Preço Ação/Ativo Ação:(+1 ....0.....-1) 
Quanto o mercado está disposto a pagar pelo Patriminio liquido de uma empresa? indica se as ações de uma empresa estão sendo negociadas com desconto (quando o P/VPA é inferior a 1) ou com prêmio (quando o P/VPA é superior a 1) em relação ao seu valor patrimonial contábil  Abaixo de 1 está abaixo do valor patrimonial. Está acima de 1 . Está acima do  Valor patrimonial.

Preço Ação/Ativo Ação:(+1 ....0.....-1)
indica se o preço da ação está refletindo adequadamente o valor dos ativos subjacentes. menor do que 1 pode indicar que o mercado está subvalorizando os ativos da empresa, o que pode ser uma oportunidade de compra. Por outro lado, um P/Ativo maior do que 1 indica que o mercado está sobrevalorizando os ativos, o que pode ser um sinal de que a ação está cara.

Preço Ação/Capital de Giro Ação O capital de giro: (Negativo ou positivo)
É a diferença entre os ativos circulantes (como caixa, contas a receber e estoques) e os passivos circulantes (como contas a pagar e empréstimos de curto prazo). Essa métrica compara o preço da ação com o capital de giro por ação, fornecendo uma medida da eficiência da gestão do capital de giro da empresa. Capital de giro negativo se refere a quando o ativo circulante não é o suficiente para pagar os credores. Quando sua empresa chega a esta situação, é necessário ficar em alerta. Claro, a curto prazo pode ser uma eventualidade, mas se isso se perpetuar por muito tempo, pode haver uma crise de liquidez.

Preço Ação/Ativos Circulantes Líquidos Ação: 
compara o preço da ação com os ativos circulantes líquidos por ação, o que pode fornecer uma medida da saúde financeira de curto prazo da empresa em relação ao preço da ação. Esse valor representa quanto sobraria de recursos caso fossem quitadas todas as obrigações da instituição naquela data.Em alguns casos, esse número pode ser negativo. Ou seja, se a empresa usasse seus ativos circulantes para pagar todas as despesas de curto e longo prazo, ainda teria dívidas a serem liquidadas.

O patrimônio líquido:
diferença entre os ativos totais e os passivos totais de uma empresa. Ele representa o valor contábil da empresa para seus acionistas, indicando o valor residual dos ativos após a liquidação de todas as dívidas. O patrimônio líquido é uma medida-chave da saúde financeira e da solidez financeira de uma empresa.

# Foco em Liquidez

São métricas financeiras que ajudam a avaliar a capacidade de uma empresa em cumprir suas obrigações de curto prazo

A liquidez corrente(+1 ....0.....-1): é uma medida da capacidade de uma empresa de pagar suas dívidas de curto prazo utilizando seus ativos circulantes, ou seja, aqueles que podem ser convertidos em dinheiro dentro de um ano. Ela é calculada dividindo os ativos circulantes pelo passivo circulante. Uma liquidez corrente maior que 1 indica que a empresa possui mais ativos circulantes do que dívidas de curto prazo, o que sugere uma posição financeira mais sólida no curto prazo. Significa que ela possui X vezes mais ativos de curto prazo do que dívidas de curto prazo.

A liquidez 2 meses é uma variação da liquidez corrente que considera uma janela de tempo mais curta para avaliar a capacidade de uma empresa em cumprir suas obrigações de curto prazo. Ela é calculada dividindo os ativos disponíveis para utilização nos próximos 2 meses pelo passivo circulante. Essa métrica fornece uma visão mais imediata da liquidez da empresa, focando em seus recursos disponíveis nos próximos dois meses.



```python
# com o uso de .copy() para garantir que df_lucro seja uma cópia do DataFrame original
df_liquidez = df_liquidez.copy()
df_liquidez.loc[:, 'Empresa'] = empresas
df_liquidez
```
